```python
import os
import re
import xlsxwriter
import pandas as pd
df=pd.read_csv(r'C:\Users\avinashexpc\Desktop\NLP\final\HCL ML Challenge\Results.csv')
outWorkbook=xlsxwriter.Workbook("C:\\Users\\avinashexpc\\Desktop\\NLP\\final\\HCL ML Challenge\\mu_res.xlsx")
outsheet=outWorkbook.add_worksheet()
outsheet.write(0,0,"Filename")
outsheet.write(0,1,"Extracted Values")
p=0
multi='C:\\Users\\avinashexpc\\Desktop\\NLP\\final\\HCL ML Challenge\\HCL ML Challenge Dataset\\'
for fil in df["Filename"]:
    with open(multi+fil+'.txt') as f:
        lines=f.readlines()
        pattern=re.compile(r'(Notes\s+)|(2020\s\s+)|(2019\s\s+)|(2018\s\s+)+.*')
        k=0
        for line in lines:
            line=line.strip()
            matches=pattern.findall(line)
            if matches!=[]:
                break
            #if(line==)
            k+=1
        if matches==[]:
            p+=1
            outsheet.write(p,0,fil)
            continue
        #print(line)
        a=line.find('2018')
        b=line.find('2019')
        c=line.find('2020')
        print(fil)
        #print(b)
        #print("****")
        d=0
        if a==-1:
            if b>c:
                d=2
            else:
                d=1
        if c==-1:
            if b>a:
                d=2
            else:
                d=1  
        pattern=re.compile(r'(([0-9a-zA-ZÂ£():|]+\s)+)\s+([0-9(),.-]+)\s+([0-9(),.-]+)')
        lst=[]
        ans={}
        with open(multi+fil+'.txt') as f:
                           lines=f.readlines()
                           i=0
                           for line in lines:
                                if line=='\n':
                                    continue 
                                matches=pattern.findall(line)
                                if matches!=[]:
                                    if matches[0][0]!='Notes ':
                                        it0=matches[0][0]
                                        it1=matches[0][2]
                                        it2=matches[0][3]
                                        it1=it1.replace("-","nan")
                                        it2=it2.replace("-","nan")
                                        it1=it1.replace("(","-")
                                        it2=it2.replace("(","-")
                                        it1=it1.replace(")","")
                                        it2=it2.replace(")","")
                                        #it0=it0.replace("Â","")
                                        it1=it1.replace(",","")
                                        it2=it2.replace(",","")
                                        matches=[(it0,matches[0][1],it1,it2)]
                                        #print(matches)
                                        ans[i]=matches
                                        lst.append(i)
                                i=i+1
       # print(ans)
        if ans=={}:
           # print("WRONG")
            p+=1
            outsheet.write(p,0,fil)
            continue
        #print("555",k)
        a=k+2
        b=max(lst)
        res={}
        abc=ans.keys()
        #print(ans,"%%%%%%%%")
        for i in range(a,b+1):
            if lines[i]=='\n':
                continue 
            if i in ans:
                res[ans[i][0][0]]=ans[i][0][d+1]
            else:
                abc=re.findall(r'(([0-9a-zA-ZÂ£:()-|]+\s)+)',lines[i])
                res[abc[0][0]]='nan'
        
        ltr="{"
        for i in res:
            ltr+='"'+str(i)[:-1]+'": '+'"'+str(res[i])+'", '
        ltr=ltr[:-2]
        ltr+="}"
        p+=1
        outsheet.write(p,0,fil)
        outsheet.write(p,1,ltr)
        print(ltr)
        print("^^^^^^^^^^^^^^^^^^^")
outWorkbook.close()

```
